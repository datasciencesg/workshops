## Working with Grammar!

Each ggplot2 visualization has 3 required components

- 1) Data - The Raw Material
- 2) **Aes**thetics - The mappings of your data to the visualization. Example map, x-axis, y-axis, fill, color
- 3) Layers - Each visualization requires at least one layer. Typically in the form of **geom**etric objects. Each layer is added using + sign!

## Visualizing Categorical Variables (Factors)

```r
# Question: What was the survival rate?

ggplot(data = titanic, aes(x = Survived)) +
    geom_bar()   #this is an example of a geom!

#There are other types of geom see: https://www.rstudio.com/wp-content/uploads/2015/03/ggplot2-cheatsheet.pdf
```

```r
#We can add layers to change how the visualization looks!

ggplot(data = titanic, aes(x = Survived)) + #specify data and aesthetic to be mapped
    geom_bar() + #geometric object bar
    theme_bw() + #changes the theme to black and white, default has a grey background
   labs(x = "Survived?", #add x-axis label name
         y = "No. of Passengers", #add y-axis label name
         title = "Titanic Survival Rates" #add chart title
        )
```

```r
#Question: What was the survival rate by sex ?

ggplot(data = titanic, aes(x = Survived, fill = Sex)) + #specify data and aesthetic to be mapped
    geom_bar() + #geometric object bar
    theme_bw() + #changes the theme to black and white, default has a grey background
    labs(x = "Survived?", #add x-axis label name
         y = "No. of Passengers", #add y-axis label name
         title = "Titanic Survival Rates" #add chart title
        ) 
```

## Using Facets to drill-down into data

```r
options(repr.plot.width = 10, repr.plot.height = 4)

#What was the survival rate by sex and class of ticket?
#
# We can leverage facets to further segment the data and enable
# "visual drill-down" into the data.
#

ggplot(data = titanic, aes(x = Sex, fill = Survived)) + #specify data and aesthetic to be mapped
    geom_bar() + #geometric object bar
    theme_bw() + #changes the theme to black and white, default has a grey background
    labs(x = "Passenger Class", #add x-axis label name
         y = "No. of Passengers", #add y-axis label name
         title = "Survival Rates by Pclass" #add chart title
        ) +
    facet_wrap(~Pclass) # use the tilda (~) to indicate factor to be included as another dimension 
```

## Visualizing Continuous Data (Histogram & Density)

```r
#Question: What is the distribution of passenger ages?

ggplot(titanic, aes(x = Age)) +
  theme_bw() +
  geom_histogram(binwidth = 5) +
  labs(y = "Passenger Count",
       x = "Age (binwidth = 5)",
       title = "Titanic Age Distribution")

#Note the Warning Message. What does it mean?
```

## Exercise 2 - Histogram

```r
#From the above histogram, how can we enhance it to view (using colors) distribution by age and survival rate?
ggplot(titanic, aes(x = Age, fill = Survived)) +
  theme_bw() +
  geom_histogram(binwidth = 5) +
  labs(y = "Passenger Count",
       x = "Age (binwidth = 5)",
       title = "Titanic Age Distribution")
```

```r
#Can we change it to a smooth density plot?
ggplot(titanic, aes(x = Age, fill = Survived)) +
  theme_bw() +
  labs(y = "Passenger Count",
       x = "Age (binwidth = 5)",
       title = "Titanic Age Distribution") +
  geom_density(alpha=0.5) #alpha from a range of 0 to 1 adjust the transparency level
```

```r
# Question: What is the survival rates by age when segmented by sex and class of ticket?
options(repr.plot.width = 8, repr.plot.height = 4)

ggplot(titanic, aes(x = Age, fill = Survived)) +
  theme_bw() +
  facet_wrap(Sex ~ Pclass) +
  geom_density(alpha = 0.5) +
  labs(y = "Age",
       x = "Survived",
       title = "Survival Rates by Age, Pclass and Sex")
```

## Measure relationship between continuous variables (geom_smooth & geom_point)

```r
options(repr.plot.width = 4, repr.plot.height = 4)
ggplot(titanic, aes(x=Age, y=Fare)) +
    geom_point()

#However, the fare would likely differ by PClass!
```

```r
ggplot(titanic, aes(x=Age, y=Fare, color = Pclass)) + #include color for easily visibility
    geom_point(alpha = 0.3) #adjust alpha to increase visibility
```

```r
#Can we add a regression line (Linear Model) and see the plots separately by Pclass?

options(repr.plot.width = 8, repr.plot.height = 4)
ggplot(titanic, aes(x=Age, y=Fare, color = Pclass)) +
    geom_point(alpha = 0.3) +
    geom_smooth(method = lm, se= FALSE) +
    facet_wrap(~Pclass)

#Problem!
#The scale makes it hard to analyse. We need the scales to be free to adjust based on each graph! solution: scale = free
```

```r
#Correct for scales = free in facet_wrap + aspect ratio = 1 so that plot is square shaped


options(repr.plot.width = 10, repr.plot.height = 4)
ggplot(titanic, aes(x=Age, y=Fare, color = Pclass)) +
    geom_point(alpha = 0.3) +
    geom_smooth(method = lm, se= FALSE) +
    facet_wrap(~Pclass, scales = "free") +
    theme(aspect.ratio=1) +
    labs(title = "Relationship between Fare and Age by Pclass")
```
